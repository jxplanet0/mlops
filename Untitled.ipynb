```python
import time
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By

url = "https://www.blibli.com/cari/laptop"
driver = webdriver.Chrome()
driver.get(url)

data = []

for i in range(1):
    WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.CSS_SELECTOR, "#blibliApp")))
    time.sleep(2)
    for j in range(20):
        time.sleep(1)
        driver.execute_script("window.scrollBy(0, 250)")
        time.sleep(1)

    driver.execute_script("window.scrollBy(50, 0)")
    time.sleep(1)

    soup = BeautifulSoup(driver.page_source, "html.parser")

    for item in soup.findAll("div", class_ = "product columns product__container__side"):
        nama_produk = item.find("div", class_ = "blu-product__name").text
        harga = item.find("div", class_ = "blu-product__price-after").text

        data.append(
            (nama_produk, harga)
        )

    time.sleep(2)
    driver.find_element(By.CSS_SELECTOR, "a[class^='blu-paging__next']").click()
    time.sleep(3)

df = pd.DataFrame(data, columns = ["Nama Barang", "Harga"])
print(df)

df.to_csv("data_blibli.csv", index = False)
print("Data Telah Tersimpan")

driver.close()
```

```python
from pycaret.datasets import get_data
from pycaret.classification import *

data = get_data('hepatitis')

exp1 = setup(data, target='Class')

imputation_strategies = ['drop', 'mean', 'median', 'mode', 'knn']

results = {}

for strategy in imputation_strategies:
    print(f"Running setup with imputation strategy: {strategy}")
    exp = setup(data, target='Class', imputation_type='simple', numeric_imputation=strategy)
    best_model = compare_models()
    results[strategy] = best_model
    
for strategy, model in results.items():
    print(f"Imputation strategy: {strategy}, Best model: {model}")

iterative_imputation_methods = [
    'lr', 'lasso', 'ridge', 'en', 'lar', 'llar', 'omp', 'br', 'ard', 'par', 
    'ransac', 'tr', 'huber', 'kr', 'svm', 'knn', 'dt', 'rf', 'et', 'ada', 
    'gbr', 'mlp', 'xgboost', 'lightgbm', 'catboost', 'dummy'
]

results_iterative = {}

for method in iterative_imputation_methods:
    print(f"Running setup with iterative imputation method: {method}")
    exp = setup(data, target='Class', imputation_type='iterative', numeric_iterative_imputer=method)
    best_model = compare_models()
    results_iterative[method] = best_model
    
for method, model in results_iterative.items():
    print(f"Iterative imputation method: {method}, Best model: {model}")
```

```python
from pycaret.datasets import get_data
from pycaret.classification import *

data = get_data('hepatitis')

exp1 = setup(data, target='Class')

simple_imputation_strategies = ['drop', 'mean', 'median', 'mode', 'knn']

iterative_imputation_methods = [
    'lr', 'lasso', 'ridge', 'en', 'lar', 'llar', 'omp', 'br', 'ard', 'par', 
    'ransac', 'tr', 'huber', 'kr', 'svm', 'knn', 'dt', 'rf', 'et', 'ada', 
    'gbr', 'mlp', 'xgboost', 'lightgbm', 'catboost', 'dummy'
][:8]

results = {}

for strategy in simple_imputation_strategies:
    print(f"Running setup with simple imputation strategy: {strategy}")
    exp = setup(data, target='Class', imputation_type='simple', numeric_imputation=strategy)
    best_model = compare_models()
    results[f"simple_{strategy}"] = best_model

for method in iterative_imputation_methods:
    print(f"Running setup with iterative imputation method: {method}")
    exp = setup(data, target='Class', imputation_type='iterative', numeric_iterative_imputer=method)
    best_model = compare_models()
    results[f"iterative_{method}"] = best_model

for param, model in results.items():
    print(f"{param}, Best model: {model}")
```
